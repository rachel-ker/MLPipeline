## Best Precision and Recall at 5%


```python
df[df['precision_at_5'] == df['precision_at_5'].max()][['model_type', 'parameters', 'trainsetlabel','precision_at_5','recall_at_5']]
```

```python
df[df['recall_at_5'] == df['recall_at_5'].max()][['model_type', 'parameters', 'trainsetlabel','precision_at_5','recall_at_5']]
```

```python
# Evaluate models at 5%

def sort_models(data, metric, top_k):
    sort = data.sort_values(metric, ascending=False)
    return sort[['model_type', 'parameters', 'trainsetlabel','precision_at_5','recall_at_5']][:top_k]

sort_models(trainset1, 'precision_at_5', 10)
```

```python
sort_models(trainset2, 'precision_at_5', 10)
```
