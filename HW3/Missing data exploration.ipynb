# Data Exploration for Missing Data

```python
import etl, pipeline
import hw3
```

```python
df = etl.read_csvfile("data/projects_2012_2013.csv")
df = etl.replace_dates_with_datetime(df, ['date_posted', 'datefullyfunded'])
y_col = "notfullyfundedin60days"
df = hw3.create_label(df, y_col)
df.columns
```

```python
etl.check_missing(df)
```

## School Metro

```python
df.school_metro.unique()
```

```python
missing_metro = df[df['school_metro'].isnull()]
missing_metro
```

```python
missing_metro.describe()[['total_price_including_optional_support', 'students_reached', 'notfullyfundedin60days']]
```

```python
df.describe()[['total_price_including_optional_support', 'students_reached', 'notfullyfundedin60days']]
```

```python
df.groupby('school_metro').agg('median')[['total_price_including_optional_support', 'students_reached', 'notfullyfundedin60days']]
```

```python
df.groupby('school_metro').agg('mean')[['total_price_including_optional_support', 'students_reached', 'notfullyfundedin60days']]
```
