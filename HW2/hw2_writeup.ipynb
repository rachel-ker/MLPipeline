## Pre-processing Data

Before building a classifier, I replace missing values with the mean and check that there are no missing values in the dataset.

```python
data = pl.replace_missing_with_mean(data)
pl.check_missing(data)
```

## Feature Generation

In this section, I create features to be used in the building of the classifier. 
Specifically, for continuous variables, we can discretize them into categorical variables. 
For the categorical variables, we can also create dummy variables for them in the model.

For this dataset, I discretize the variables 'RevolvingUtilizationOfUnsecuredLines' and 'DebtRatio' to \[0,1) and [1,+Inf). This makes sense because both of these variables are a ratio. When either of the ratio is [0,1), it tells us that the balance of credits is less than the sum of credit limits or that the debt is less than the monthly gross income. Otherwise, it exceeds.

```python
data = pl.discretize(data, 'RevolvingUtilizationOfUnsecuredLines', [0,1])
data = pl.discretize(data, 'DebtRatio', [0,1])
data.columns
```

For all categorical variables in the dataset, I convert them to categorical variables into dummy variables since decision trees require numerical inputs.

```python
data = pl.create_dummies(data, 'RevolvingUtilizationOfUnsecuredLines_discrete')
data = pl.create_dummies(data, 'DebtRatio_discrete')
data = pl.create_dummies(data, 'zipcode')
```

```python
data.columns
```

## Build and Evaluate Classifier

To build the decision tree for this data set, I select the relevant features to keep and train a decision tree model using the features and target variable we hope to classify on. 

We can see the trained decision tree model below that is trained on a training set containing 80% of the dataset. 

After the tree classifier is built, I test it on the other 20% of the dataset (the test set) to get an accuracy score by comparing the predicted labels with the actual labels. As we can see, the accuracy of the prediction of the tree visualized below is about 87%.

```python
data_features = pl.select_features(data, 'SeriousDlqin2yrs', ['RevolvingUtilizationOfUnsecuredLines',
                                  'age', 'NumberOfTime30-59DaysPastDueNotWorse', 'DebtRatio',
                                  'MonthlyIncome', 'NumberOfOpenCreditLinesAndLoans',
                                  'NumberOfTimes90DaysLate', 'NumberRealEstateLoansOrLines',
                                  'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents',
                                  'RevolvingUtilizationOfUnsecuredLines_discrete_0_to_1',
                                  'RevolvingUtilizationOfUnsecuredLines_discrete_1_to_22001.0',
                                  'DebtRatio_discrete_0_to_1', 'DebtRatio_discrete_1_to_106886.0',
                                  'zipcode_60601', 'zipcode_60618', 'zipcode_60625', 'zipcode_60629',
                                  'zipcode_60637', 'zipcode_60644'])
```

```python
pl.build_and_test_classifier(data_features, 'SeriousDlqin2yrs', 0.2, max_depth=4, min_leaf=1000)
```
