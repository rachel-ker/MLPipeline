# Assignment 4 - Clustering

Loading Data

```python
import config, etl
data = etl.read_csvfile(config.DATAFILE)
```

Data Cleaning

```python
# Data Cleaning
data = etl.replace_dates_with_datetime(data, config.TO_DATE)
data = etl.replace_missing_with_mode(data, data, config.MISSING)
```

```python
data = data[config.FEATURES]
data.head()
```

```python
# Create Dummies
for cat in config.CATEGORICAL:
    data = etl.create_dummies(data, cat)
```

Build Clusters

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

clusters = [2]
for n in clusters:
    kmeans = KMeans(n_clusters=n, random_state=0).fit(data)
    data['pred_label'] = pd.Series(kmeans.labels_)
    print(data.groupby('pred_label').mean())
```
